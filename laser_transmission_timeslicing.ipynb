```python
def plot_I_biggest_nodes():
    import pandas as pd
    import matplotlib.pyplot as plt

    # Load the CSV file into a DataFrame
    df = pd.read_csv("output_csv/I.csv", header=None)

    # Find the 5 rows with the largest initial values (first column)
    top_5_rows = df.nlargest(5, 0)  # Select the top 5 rows based on the first column

    # Plot the time series for these 5 rows
    plt.figure(figsize=(10, 6))
    for index, row in top_5_rows.iterrows():
        plt.plot(row, label=f"Location {index}")

    # Customize the plot
    plt.title('Top 5 Locations with Largest Initial Infected Populations')
    plt.xlabel('Day')
    plt.ylabel('Infected Population')
    plt.legend(title="Locations")
    plt.grid(True)

    # Show the plot
    plt.show()
plot_I_biggest_nodes()
```
